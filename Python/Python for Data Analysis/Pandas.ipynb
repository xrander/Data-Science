### Making Visualizations in Pandas

#### Common aguments
- kind
- xlabel

#### To save the plots
we import matplotlib.pyplot


```python
import matplotlib.pyplot as plt
```

```python
df_pivot_pop.plot(kind='line', xlabel='Year', ylabel='Population', title = 'Population (1955 - 2020)', figsize=(10, 6))

#saving the plot
plt.savefig('line_plot.jpeg')
#Showing plot alone
plt.show()
```

```python
df_pivot_pop.index
```

```python
df_pivot_pop_2020 = df_pivot_pop[df_pivot_pop.index.isin([2020])]
#the .isin method is used to filter and make selection
```

```python
df_pivot_pop_2020
```

```python
df_pivot_pop_2020=df_pivot_pop_2020.T
#.T transposes the data
```

```python
df_pivot_pop_2020.plot(kind='bar',xlabel='Year', ylabel='Population', title = 'Population (2020)', figsize=(10, 6), color = 'purple')
plt.savefig('my_test_bar_plot.jpeg')
#Showing plot alone
plt.show()
```
